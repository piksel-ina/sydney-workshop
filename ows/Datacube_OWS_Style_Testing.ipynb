```python
# AOI over jakarta
lon = [106.655, 106.855]
lat = [-6.255, -6.055]
time = ["2025-01-01", "2025-12-31"]

dc = datacube.Datacube()
data = dc.load(
    product="geomad_s2_annual",
    lon=lon,
    lat=lat,
    time=time,
    dask_chunks={},
    group_by="solar_day",
    output_crs="6933",
    resolution=20,
)
data
```

```python
subset = data.squeeze().compute()
```

```python
subset.odc.explore(vmin=0, vmax=3000, bands=["red", "green", "blue"])
```

```python
# NDVI - Normalized Difference Vegetation Index
style = {
    "name": "ndvi",
    "title": "NDVI - Red, NIR",
    "abstract": "Normalised Difference Vegetation Index - a derived index that correlates well with the existence of vegetation",
    "index_function": {
        "function": "datacube_ows.band_utils.norm_diff",
        "mapped_bands": True,
        "kwargs": {"band1": "nir", "band2": "red"},
    },
    "needed_bands": ["red", "nir"],
    "color_ramp": [
        {"value": -0.0, "color": "#8F3F20", "alpha": 0.0},
        {"value": 0.0, "color": "#8F3F20", "alpha": 1.0},
        {"value": 0.1, "color": "#A35F18"},
        {"value": 0.2, "color": "#B88512"},
        {"value": 0.3, "color": "#CEAC0E"},
        {"value": 0.4, "color": "#E5D609"},
        {"value": 0.5, "color": "#FFFF0C"},
        {"value": 0.6, "color": "#C3DE09"},
        {"value": 0.7, "color": "#88B808"},
        {"value": 0.8, "color": "#529400"},
        {"value": 0.9, "color": "#237100"},
        {"value": 1.0, "color": "#114D04"},
    ],
    "legend": {
        "begin": 0.0,
        "end": 1.0,
        "ticks": [0.0, 0.5, 1.0],
    },
}
```

```python
plot_image_with_style_cfg(style, subset)
```
